```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Data Collection
# A dataset in CSV format
data = pd.read_csv('VCT_2024.csv')
```

```python
# 2. Data Preprocessing
# Check the first few rows to understand data
data.head()
# Normalize data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['ACS', 'K/D Ratio', 'Headshot Percentage']] = scaler.fit_transform(data[['ACS', 'K/D Ratio', 'Headshot Percentage']])
# Encoding categorical data
data = pd.get_dummies(data, columns=['Team'], drop_first=True)
# Separating features and target
X = data.drop('Win', axis=1)  # 'Win' is the target column
y = data['Win']
# Splitting dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 3. Model Training
# SVM
svm_model = SVC(kernel='linear', C=1)
svm_model.fit(X_train, y_train)
# Random Forest
rf_model = RandomForestClassifier()
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30]
}
grid_search = GridSearchCV(rf_model, param_grid, cv=3)
grid_search.fit(X_train, y_train)
best_rf_model = grid_search.best_estimator_
```
